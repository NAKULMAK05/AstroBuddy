```python
import requests
import json
import plotly.graph_objects as go

# Base URL for the Flask API
BASE_URL = "https://ac52-34-125-55-160.ngrok-free.app"

# 1. Call to `/api/birth-chart` endpoint
def get_birth_chart():
    # Take user input for all required fields
    birth_date = input("Enter your birth date (YYYY-MM-DD): ")
    birth_time = input("Enter your birth time (HH:MM in 24-hour format): ")
    birth_lat = float(input("Enter your birth latitude: "))
    birth_lon = float(input("Enter your birth longitude: "))
    phone_number = input("Please enter your phone number: ")

    payload = {
        "birth_date": birth_date,
        "birth_time": birth_time,
        "birth_lat": birth_lat,
        "birth_lon": birth_lon,
        "phone_number": phone_number
    }

    url = f"{BASE_URL}/api/birth-chart"
    response = requests.post(url, json=payload)

    if response.status_code == 200:
        birth_chart_data = response.json()  # Get birth chart data as JSON

        # Store the data in Astra DB (make sure to connect to your database first)
        collection = db.get_collection('spiritual_ai')
        collection.insert_one(birth_chart_data)

        print("Birth chart data stored successfully.")

        return birth_chart_data  # Optionally return the modified data
    else:
        print(f"Error: {response.status_code}, {response.text}")
        return None

# 2. Call to `/api/house-chart` endpoint
def get_house_chart():
    url = f"{BASE_URL}/api/house-chart"
    response = requests.post(url)  # No parameters required for this endpoint
    if response.status_code == 200:
        return response.json()  # Return the Plotly chart JSON
    else:
        print(f"Error: {response.status_code}, {response.text}")
        return None

# Example Usage in the Notebook
birth_chart_data = get_birth_chart()
if birth_chart_data:
    print("Birth Chart Data:", json.dumps(birth_chart_data, indent=4))

house_chart_json = get_house_chart()
if house_chart_json:
    # Create and render the Plotly chart from JSON
    fig = go.Figure(house_chart_json)
    fig.show()
```
